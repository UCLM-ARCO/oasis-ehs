# Panel generation
This notebook generates panel sizes (width and height in cm) and saves the result to a CSV named `panels.csv`.

```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
# Generate unique unordered pairs from 1 cm x 1 cm to 30 cm x 30 cm (width <= height)
min_size = 1  # cm
max_size = 15  # cm

widths = np.arange(min_size, max_size + 1, 1)
heights = np.arange(min_size, max_size + 1, 1)
pairs = [(w, h) for w in widths for h in heights if w <= h]
df_grid = pd.DataFrame(pairs, columns=['width_cm', 'height_cm'])
print(f'Generated {len(df_grid)} unique panels (width <= height)')
csv_grid = 'panel_areas.csv'
df_grid.to_csv(csv_grid, index=False)
print(f'Grid saved to: {csv_grid}')
```
